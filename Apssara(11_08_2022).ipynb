4)Explanation:
Bar chart: It shows relationship between numerical and categorical variables.The height of the bar chart cooresponds to the value.

Histogram:It is a graphical representation that displays data by way of bars to show the frequency of numerical data.

Scatter plot:It shows relationship between two variables and is represented by dots.The scatter()allows to create a scattered plot.

```python
score_df = pd.read_csv('/content/drive/My Drive/score.csv')
data = score_df
display(data)
```

```python
plt.scatter(data['Hours'], data['Scores'])
plt.title("Scatter Plot")
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import io
plt.bar(data['Hours'], data['Scores'])
plt.title("Bar Chart")
plt.xlabel('Hours' )
plt.ylabel('Scores')
plt.show()
```

```python
plt.hist(data['Hours'])
plt.title("Histogram")
plt.show()
```
